## Import Semua Packages/Library yang Digunakan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
import os
```

## Data Wrangling

### Gathering Data

```python
# Set the directory containing the CSV files
directory = "Dataset/"

# Initialize an empty dictionary to store DataFrames
dfs = {}

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        # Extract the city name from the filename
        district = filename.split("_")[2]
        district = district.split(".")[0]
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(os.path.join(directory, filename))
        dfs[district] = df
```

```python
Aoti_df = dfs['Aotizhongxin']
Chan_df = dfs['Changping']
Ding_df = dfs['Dingling']
Dong_df = dfs['Dongsi']
Guan_df = dfs['Guanyuan']
Guch_df = dfs['Gucheng']
Huai_df = dfs['Huairou']
Nong_df = dfs['Nongzhanguan']
Shun_df = dfs['Shunyi']
Tian_df = dfs['Tiantan']
Wanli_df = dfs['Wanliu']
Wansh_df = dfs['Wanshouxigong']
```

```python
# Calculate the percentage of missing values for each column separately for each station
missing_percentage_station = combined_df.groupby('station').agg(lambda x: x.isna().mean() * 100)

# Print the result
print("Percentage of missing values for each column for each station:")
print(missing_percentage_station)
```

**Since the percentage of is not that significant (<= 6%), we can assume the missing value is the mean of each column in each station. This is done to preserve integrity of data, so we can explore daily data with no missing days**

### Cleaning Data

```python
# Filter float type columns
float_columns = combined_df.select_dtypes(include='float64').columns

# Calculate the mean for the float type columns grouped by 'station'
station_means = combined_df.groupby('station')[float_columns].mean()

# Print the mean values for each station
print("Mean values for float type columns grouped by each station:")
print(station_means)
```

```python
# Convert 'year', 'month', 'day', and 'hour' columns to datetime format and insert at first position
combined_df.insert(1, 'date', pd.to_datetime(combined_df[['year', 'month', 'day', 'hour']]))

# Print the updated DataFrame
print(combined_df)
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Define the aggregation functions for the columns
aggregation_functions = {
    'PM2.5': 'mean',
    'PM10': 'mean',
    'SO2': 'mean',
    'NO2': 'mean',
    'CO': 'mean',
    'O3': 'mean',
}

# Group by 'station' and 'month', then aggregate the columns using the defined functions
grouped_df = combined_df.groupby(['station', 'month']).agg(aggregation_functions)

# Print the grouped DataFrame
print(grouped_df)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Select the pollutant columns using the specified column indices
pollutant_columns = combined_df.iloc[:, 5:11]

print(pollutant_columns)
```

```python
# Define figure size and number of columns for subplots
plt.figure(figsize=(15, 20))
num_cols = 3

# Iterate over each column except 'date' and 'station'
pollution_cols = [col for col in combined_df.columns if col not in ['date', 'station']]
num_plots = len(pollution_cols)

# Plot each pollution value
for i, col in enumerate(pollution_cols, start=1):
    plt.subplot((num_plots // num_cols) + 1, num_cols, i)
    combined_df[col].plot()
    plt.title(f'Time Series Plot of {col}')
    plt.xlabel('Date')
    plt.ylabel(f'{col} concentration')
    plt.grid(True)

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
